```python
import pandas as pd
import numpy as np

NBA = pd.read_csv('assignment4/assets/nba.csv')
NFL = pd.read_csv('assignment4/assets/nfl.csv')
NHL = pd.read_csv('assignment4/assets/nhl.csv')
MLB = pd.read_csv('assignment4/assets/mlb.csv')
```

```python
NBA=NBA[NBA['year']==2018]
NBA['team']=NBA['team'].str.replace(r"[*]","")
NBA['team']=NBA['team'].str.replace(u'\xa0', u' ')
NBA['team']=NBA['team'].str.replace(r"[ ]*\([0-9]*\)","")
NBA['W']=pd.to_numeric(NBA['W'])
NBA['L']=pd.to_numeric(NBA['L'])
NBA['W-L%']=(NBA['W'])/(NBA['W']+NBA['L'])
NBA=NBA[['year','League','team','W','L','W-L%']]
NBA.head()
```

```python
NFL=NFL[NFL['year']==2018]
NFL=NFL.sort_values("DSRS",ascending=False)
NFL=NFL.iloc[8:40]
NFL['team']=NFL['team'].str.replace(r"[+*]","")
NFL['W']=pd.to_numeric(NFL['W'])
NFL['L']=pd.to_numeric(NFL['L'])
NFL['W-L%']=(NFL['W'])/(NFL['W']+NFL['L'])
NFL=NFL[['year','League','team','W','L','W-L%']]
NFL.head()
```

```python
NHL=NHL[NHL['year']==2018]
NHL=NHL.sort_values("SRS",ascending=False)
NHL=NHL.iloc[4:36]
NHL['team']=NHL['team'].str.replace(r"[*]","")
NHL['W']=pd.to_numeric(NHL['W'])
NHL['L']=pd.to_numeric(NHL['L'])
NHL['W-L%']=(NHL['W'])/(NHL['W']+NHL['L'])
NHL=NHL[['year','League','team','W','L','W-L%']]
NHL.head()
```

```python
MLB=MLB[MLB['year']==2018]
MLB['W']=pd.to_numeric(MLB['W'])
MLB['L']=pd.to_numeric(MLB['L'])
MLB['W-L%']=(MLB['W'])/(MLB['W']+MLB['L'])
MLB=MLB[['year','League','team','W','L','W-L%']]
MLB.head()
```
