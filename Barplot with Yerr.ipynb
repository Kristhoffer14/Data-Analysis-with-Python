```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=['1992','1993','1994','1995'])
df
```

```python
df=df.T
df.describe()
```

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from random import randint
```

```python
#Observando los datos con un BoxPlot
plt.figure()
plt.boxplot([df['1992'],df['1993'],df['1994'],df['1995']],whis='range')
```

```python
#Creando los datos resumidos (sumatoria de valores por año; error estandar por año)
sum92=np.sum(df['1992'])
sum93=np.sum(df['1993'])
sum94=np.sum(df['1994'])
sum95=np.sum(df['1995'])

stderr_92 = np.std (df['1992'], ddof = 1) / np.sqrt (len (df['1992']))
stderr_93 = np.std (df['1993'], ddof = 1) / np.sqrt (len (df['1993']))
stderr_94 = np.std (df['1994'], ddof = 1) / np.sqrt (len (df['1994']))
stderr_95 = np.std (df['1995'], ddof = 1) / np.sqrt (len (df['1995']))
```

```python
#Graficando las barras con un color diferente cada uno
plt.figure()

x=range(len(df.columns))
y=[sum92,sum93,sum94,sum95]


linear_err=[stderr_92*10000,stderr_93*10000,stderr_94*10000,stderr_95*10000]

bars = plt.bar(x,y,width=0.99,yerr=linear_err,capsize=20)

#Colors
bars[0].set_color('#063F75')
bars[1].set_color('#FFAF88')
bars[2].set_color('#9FE4FC')
bars[3].set_color('#781515')


plt.xticks(x,df.columns,alpha=0.8)

plt.show()

#y2=[1.65*100000000,1.65*100000000,1.65*100000000,1.65*100000000]
#plt.plot(x,y2,'-',c='b')
```
